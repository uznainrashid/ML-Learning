### **Scope of Claims: What We Can and Cannot Claim**

To maintain scientific integrity and align with FlyRank's standards, we establish clear boundaries on what our analysis and model can and cannot claim:

#### **What We CAN Claim:**
* **Observational Patterns:** We can identify which observable features (like age, CTR, and word count) are historically associated with traffic decline.
* **Directional Trends:** We can show the direction of movement (e.g., "stale pages have a higher probability of dropping in visibility").
* **Decision-Support Prioritization:** We can provide a ranked list of recommendations to help human editors utilize their limited capacity more efficiently.

#### **What We CANNOT Claim:**
* **Causal Proof:** We cannot prove that a low word count or old age *caused* the decline. To prove causality, we would need controlled A/B testing, which this observational dataset does not support.
* **Predicting Google's Algorithm:** We are not predicting or reverse-engineering Google's ranking algorithms. We are simply modeling historical patterns in our own tracking data.
* **Guaranteed Recovery:** We cannot guarantee that updating/refreshing a high-scoring page will definitely recover its traffic. Our model only flags it as a strong *opportunity* for review.
